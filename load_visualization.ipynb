## Load visualization for zone/zones

#### Specify the year, zone

```python
year = 2040
zones = ["NY_A_West", "NY_B_Genesee"]
```

#### Shapes

```python
df = pd.DataFrame(
        {'timestamp': pd.date_range('{}-01-01'.format(year), '{}-01-01'.format(year+1), freq='1H', closed='left')}
     )
```

```python
# Day of week
# Monday = 0, Sunday = 6
df['month'] = df['timestamp'].dt.month
df['hour'] = df['timestamp'].dt.hour+1
df['week'] = df['timestamp'].dt.weekday
df
```

```python
# select the corresponding baseload shape based on the start of the year. 
row_start = df.week[0]*24
row_end = row_start+len(df.week)
for zone in zones: 
    df['Base_shape_{}'.format(zone)]=df_baseload_shape[zone][row_start:row_end].to_numpy() # turn data into np series to get rid of default index
df
```

```python
# hourly electrification shape
elecload_shape = ['SH_shape', 'WH_shape', 'HDV_shape', 'LDVU_shape', 'LDVM_shape']
for zone in zones: 
    for i in elecload_shape: 
        if len(df)==8760: 
            df["{}_{}".format(i, zone)] = df_elecload_shape[df_specified_area.loc[zone, i]]
        else:
            df["{}_{}".format(i, zone)] = df_elecload_shape[df_specified_area.loc[zone, i]]
            df["{}_{}".format(i, zone)][-24:] = df_elecload_shape[df_specified_area.loc[zone, i]][-24:]
```

#### Load

```python
# baseload = baseload shape * hourly average load (MW)
for zone in zones:
    df['Base_load_{}'.format(zone)] = df['Base_shape_{}'.format(zone)] * df_baseload.loc[pd.IndexSlice[year, 13], zone]
```

```python
# hourly electrification load 
elec_type = ['SH', 'WH', 'HDV','LDVU', 'LDVM']
for elec_type in elec_type:
    for zone in zones:
        capacity = df_elecload.loc[pd.IndexSlice[zone, elec_type], str(year)] * (-1)
        df['{}_load_{}'.format(elec_type, zone)] = df["{}_shape_{}".format(elec_type, zone)] * capacity
```

#### Month-Hour Average of Load

```python
# get the month-hour average of load 
df_load = df.copy().groupby(['month', 'hour']).mean()
df_load = df_load.loc[:,'Base_load_{}'.format(zones[0]):"LDVM_load_{}".format(zones[-1])]
df_load = df_load.reset_index()
df_load
```

```python
fig, axs = plt.subplots(len(zones), 12, sharex=True, sharey=True, figsize=(20, 5))

for j in range(0, len(zones)): 
    for i in df_load.month.unique():
        df3 = df_load[df_load.month==i]

        x = df3['hour']
        y = [df3['Base_load_{}'.format(zones[j])], 
             df3['SH_load_{}'.format(zones[j])],
             df3['WH_load_{}'.format(zones[j])],
             df3['HDV_load_{}'.format(zones[j])],
             df3['LDVU_load_{}'.format(zones[j])], 
             df3['LDVM_load_{}'.format(zones[j])]]

        axs[j, i-1].stackplot(x, y, labels=['Base_load', 'SH_load','WH_load','HDV_load','LDVU_load','LDVM_load'])
        axs[j, i-1].title.set_text('Month {}'.format(i))
        axs[j, i-1].set_xticks([6, 12, 18])
    axs[j, 0].set_ylabel('MWh')
    axs[j, 11].legend(loc='lower right')

# handles, labels = axs.get_legend_handles_labels()
# fig.legend(handles, labels, loc='lower_right')
fig.suptitle('Month-hour Average of Load in {}'.format(zones))
```

#### Annual Load

```python
zones = ['NY_J_NYC']
# zones = [zone for zone in df_baseload.columns if "NY" in zone]
```

```python
df_annual_load = pd.DataFrame({'Year':range(2017, 2051), 'Base_load':0, 'SH_load':0, 'WH_load':0,'HDV_load':0,'LDV_load':0})
df_annual_load = df_annual_load.set_index('Year')

for zone in zones: 
    df_annual_load['Base_load'] += df_base_load_by_area.loc[zone].values
    df_annual_load['HDV_load'] += df_hdv_load_by_area.loc[zone].values
    df_annual_load['LDV_load'] += df_ldv_load_by_area.loc[zone].values
    df_annual_load['SH_load'] += df_sh_load_by_area.loc[zone].values
    df_annual_load['WH_load'] += df_wh_load_by_area.loc[zone].values

# turn load in MWh into TWh
df_annual_load = df_annual_load/(10**6)
```

```python
x = df_annual_load.index.values
y = [df_annual_load['Base_load'], df_annual_load['SH_load'], df_annual_load['WH_load'], df_annual_load['HDV_load'], df_annual_load['LDV_load']]

plt.figure(figsize=(10,5))
plt.stackplot(x, y, labels= df_annual_load.columns)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=5)
plt.ylabel('TWh')
plt.title('Load Forecast')
```

## Output summary statistics for zone/zones

```python
year = 2020
# zones = ["NY_A_West", "NY_B_Genesee"]
# zones = ['NY_A_West',
#  'NY_B_Genesee',
#  'NY_C_Central',
#  'NY_D_North',
#  'NY_E_MohawkValley',
#  'NY_F_Capital',
#  'NY_G_HudsonValley',
#  'NY_H_Millwood',
#  'NY_I_Dunwoodie',
#  'NY_J_NYC',
#  'NY_K_LongIsland']
zones = [zone for zone in df_baseload.columns if "NY" in zone]
```

```python
df_all = pd.DataFrame(
        {'timestamp': pd.date_range('{}-01-01'.format(year), '{}-01-01'.format(year+1), freq='1H', closed='left')}
     )
df_all['month'] = df_all['timestamp'].dt.month
df_all['hour'] = df_all['timestamp'].dt.hour
df_all['week'] = df_all['timestamp'].dt.weekday
df_all['System_Total_Load'] = 0

for zone in zones:
    df_shapes = df_all.copy()
    
    # select the corresponding baseload shape based on the start of the year. 
    row_start = df_shapes.week[0]*24
    row_end = row_start+len(df_shapes.week)
    
    df_shapes['Base_shape']=df_baseload_shape[zone][row_start:row_end].to_numpy() # turn data into np series to get rid of default index

    # hourly electrification shape
    elecload_shape = ['SH_shape', 'WH_shape', 'HDV_shape', 'LDVU_shape', 'LDVM_shape']
    for i in elecload_shape: 
        if len(df)==8760: 
            df_shapes[i] = df_elecload_shape[df_specified_area.loc[zone, i]]
        else:
            df_shapes[i] = df_elecload_shape[df_specified_area.loc[zone, i]]
            df_shapes[i][-24:] = df_elecload_shape[df_specified_area.loc[zone, i]][-24:]
            
    # baseload = baseload shape * hourly average load (MW)
    df_all[str(zone+'_Base_load')] = df_shapes['Base_shape'] * df_baseload.loc[pd.IndexSlice[year, 13], zone]
    
    # hourly electrification load 
    elec_type = ['SH', 'WH', 'HDV','LDVU', 'LDVM']
    for elec_type in elec_type:
        capacity = df_elecload.loc[pd.IndexSlice[zone, elec_type], str(year)] * (-1)
        df_all[str(zone+"_"+elec_type+'_load')] = df_shapes[str(elec_type+'_shape')] * capacity
        
    df_all[str(zone+'_Total_load')] = df_all[str(zone+'_Base_load')] + df_all[str(zone+'_SH_load')] + df_all[str(zone+'_WH_load')] + df_all[str(zone+'_HDV_load')] + df_all[str(zone+'_LDVU_load')] + df_all[str(zone+'_LDVM_load')]
    
    df_all['System_Total_Load'] = df_all['System_Total_Load'] + df_all[str(zone+'_Total_load')]
```

```python
system_peak_time = df_all['timestamp'].iloc[df_all.System_Total_Load.argmax()]

system_peak_row = df_all.System_Total_Load.argmax()

co_peak = []
non_peak = []
for zone in zones:
    co_peak.append(df_all[str(zone+'_Total_load')].iloc[system_peak_row])
    non_peak.append(df_all[str(zone+'_Total_load')].max())
summary_table = {"Zones":zones, "Coincident_Peak":co_peak, "Non-coincident_Peak": non_peak}
df_summary_table = pd.DataFrame.from_dict(summary_table)
```
